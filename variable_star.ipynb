```python
import lightkurve as lk
import matplotlib.pyplot as plt
id=input('Enter name the star whose period is to be found out :')
search_result = lk.search_lightcurve(id, mission='TESS')
if search_result:
    lk_unbin= search_result[0].download()
    print(search_result[0])
    lk_unbin = lk_unbin.remove_nans().normalize()
    lk=lk_unbin.bin(time_bin_size=0.1)
    lk.plot(marker='.',color='blue')
    plt.show()
    period=lk.to_periodogram().period_at_max_power
    print('Period=', period)
    lk.fold(period).plot()
    pg=lk.to_periodogram()
    pg.plot()
    plt.show()

else:
    print("No light curves found for", id ,"with mission TESS.")
```
